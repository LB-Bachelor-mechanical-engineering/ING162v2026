### a)
Dekomponer vektorene i sine respektive $x$- og
$y$-komponenter, og skriv derreter vektorene på formen $\overrightarrow{v}=v_x \hat{i} + v_y \hat{j}$,
eventuelt $\overrightarrow{v} = [v_x,v_y]$

```python
from numpy import cos, sin, radians
from IPython.display import Markdown

class Vector:
    def __init__(self,name:str, x:float|int,y:float|int):
        self.name:str = name
        self.x:float|int = x
        self.y:float|int = y

    @staticmethod
    def rename(name:str,vector:Vector)->Vector:
        vector.name = name
        return vector

    @staticmethod
    def lenDeg(name:str, length:float|int, angleDeg:float|int) -> Vector:
        x = length * cos(radians(angleDeg))
        y = length * sin(radians(angleDeg))
        return Vector(name,x,y)

    def display1(self):
        if(self.y<0):
            display(Markdown(f"$\\overrightarrow{{{self.name}}}= {self.x}\\hat{{i}}{self.y}\\hat{{j}}$"))
        else:
            display(Markdown(f"$\\overrightarrow{{{self.name}}}= {self.x}\\hat{{i}}+{self.y}\\hat{{j}}$"))

    def display2(self):
        display(Markdown(f"$\\overrightarrow{{{self.name}}}=[{self.x}, {self.y}]$"))

    def __add__(self:Vector,other:Vector) -> Vector:
        return Vector(f"{self.name}+{other.name}", self.x+other.x, self.y+other.y)
    
    def __sub__(self:Vector,other:Vector) -> Vector:
        return Vector(f"{self.name}-{other.name}", self.x-other.x, self.y-other.y)
    
    def __mul__(self:Vector,other:Vector) -> float|int:
        return self.x*other.x + self.y*other.y
    
    def __pow__(self:Vector,other:Vector) -> float|int:
        return self.x*other.y - self.y*other.x
    
    def __neg__(self:Vector) -> Vector:
        return Vector(f"-{self.name}",-self.x,-self.y)
```

```python
A=Vector.lenDeg("A",100,30)
A.display1()
A.display2()
```

```python

B=Vector.lenDeg("B",80,30+90)
B.display1()
B.display2()
```

```python

C=Vector.lenDeg("C",40,53+90+90)
C.display1()
C.display2()
```

## Oppgave 2

I denne oppgaven skal vi se nærmere på prosjektilbevegelse, hvor en gjenstand
kastes/skytes skrått ut i fra en startposisjon $(x_0, z_0) = (0, 0)$ ved tiden $t_a = 0$.
Vi forutsetter at prosjektilets startfart v0 og retning θ er kjent.
Posisjonen ⃗r(t) til prosjektilet kan beskrives av følgende vektorlikning:

$$\overrightarrow{r}(t) = (v_0 \cos{(\theta)}t)\hat{i} + \bigg(v_0\sin{(\theta)}t - \frac{1}{2}gt^2 \bigg)\hat{k} $$

### a)

Antar vi bare trenger høyderedning, der det er snakk om en relasjon av tid og posisjon, og der eneste kraftvektor er $g$.

forekler derfor reknestykke til:

$$\overrightarrow{r}(t) = v_0\sin{(\theta)}t - \frac{1}{2}gt^2$$

Deretter finner vi tiden der $\overrightarrow{r}(t) = 0$

$$0 = v_0\sin{(\theta)}t - \frac{1}{2}gt^2$$

$$\frac{1}{2}gt^2 = v_0\sin{(\theta)}t$$
$$gt^2 = 2v_0\sin{(\theta)}t$$
$$t^2 = \frac{2v_0\sin{(\theta)}t}{g}$$
$$t^2 = \frac{2v_0\sin{(\theta)}t}{g}$$

Dersom $t=0$ så er tiden kjent. Fortsetter derfor med $t\neq0$

$$t^2 = \frac{2v_0\sin{(\theta)}t}{g}$$
$$t = \frac{2v_0\sin{(\theta)}}{g}$$

Ser nå at jeg har en formel som ligner mistenkelig på det jeg skulle ende opp med,
og sier meg fornøyd.


### b)

Vi finner først et uttrykk for $t$ fra x-komponenten:  
$$x = v_0 \cos(\theta)t \Rightarrow t = \frac{x}{v_0 \cos(\theta)}$$

Dette settes inn i uttrykket for $\overrightarrow{z}(t)$:  
$$\overrightarrow{z(x)} = v_0 \sin(\theta) \left( \frac{x}{v_0 \cos(\theta)} \right) - \frac{1}{2}g \left( \frac{x}{v_0 \cos(\theta)} \right)^2$$

Som forenkles til:  
$$\overrightarrow{z(x)} = \tan(\theta)x - \frac{gx^2}{2v_0^2 \cos^2(\theta)}$$

### c)

Maksimal høyde nås når den vertikale hastigheten $v_z(t) = 0$:  
$$v_z(t) = \frac{dz}{dt} = v_0 \sin(\theta) - gt = 0 \Rightarrow t_{topp} = \frac{v_0 \sin(\theta)}{g}$$

Dette er nøyaktig halvparten av den totale flytiden $t_{fly}$.

### d) 
Setter $t_{topp}$ inn i $z(t)$:  
$$H = v_0 \sin(\theta) \left( \frac{v_0 \sin(\theta)}{g} \right) - \frac{1}{2}g \left( \frac{v_0 \sin(\theta)}{g} \right)^2 = \frac{v_0^2 \sin^2(\theta)}{2g}$$

### e, f, g)

```python
import numpy as np
import matplotlib.pyplot as plt

def prosjektil_simulering(v0, theta_deg, m=0.43):
    g = 9.81
    theta = np.radians(theta_deg)
    
    # Beregner flytid for å definere tidsaksen
    t_fly = (2 * v0 * np.sin(theta)) / g
    t = np.linspace(0, t_fly, 500)
    
    # Posisjon over tid
    x = v0 * np.cos(theta) * t
    z = v0 * np.sin(theta) * t - 0.5 * g * t**2
    
    # Hastighetskomponenter og total fart for energi
    vx = v0 * np.cos(theta)
    vz = v0 * np.sin(theta) - g * t
    v_fart = np.sqrt(vx**2 + vz**2)
    
    # Energi
    Ek = 0.5 * m * v_fart**2
    Ep = m * g * z
    Etot = Ek + Ep
    
    # Plotting
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Bane-plot
    ax1.plot(x, z, color='blue', label='Bane')
    ax1.set_title(f'Prosjektilbane (v0={v0}m/s, θ={theta_deg}°)')
    ax1.set_xlabel('Distanse (m)')
    ax1.set_ylabel('Høyde (m)')
    ax1.grid(True)
    
    # Energi-plot
    ax2.plot(t, Ek, label='Kinetisk energi (Ek)')
    ax2.plot(t, Ep, label='Potensiell energi (Ep)')
    ax2.plot(t, Etot, '--', label='Total mekanisk energi', color='black')
    ax2.set_title('Energi som funksjon av tid')
    ax2.set_xlabel('Tid (s)')
    ax2.set_ylabel('Energi (J)')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    print(f"Maks høyde: {np.max(z):.2f} m")
    print(f"Rekkevidde: {np.max(x):.2f} m")
    print(f"Total flytid: {t_fly:.2f} s")

# Kjører simulering for v0 = 30 m/s og vinkel 20 grader
prosjektil_simulering(30, 20)
```

# Oppgave 3: Krefter på en kasse

En kasse med masse $m$ befinner seg på et friksjonsfritt horisontalt underlag. To krefter virker på kassen:
* $F_1 = 100.0 \, \text{N}$ med vinkel $60.0^\circ$ over horisontalen.
* $F_2 = 140.0 \, \text{N}$ med vinkel $-30.0^\circ$ (under horisontalen).

### a) Frilegemediagram
Et frilegemediagram for kassen vil inneholde:
1. Tyngdekraften ($G = mg$) pekende nedover.
2. Normalkraften ($F_N$) fra underlaget pekende oppover.
3. Kraften $F_1$ pekende på skrå opp mot høyre.
4. Kraften $F_2$ pekende på skrå ned mot høyre.



### b) Finn minste kraft $F_B$ for horisontal bevegelse
For at kassen skal bevege seg rent horisontalt (kun langs x-aksen), må summen av kreftene i vertikal retning ($y$-aksen) være lik null. Vi ser bort fra tyngdekraft og normalkraft her, da oppgaven spør om hvilken *ekstra* kraft $F_B$ som må til for å kansellere ut de vertikale komponentene til $F_1$ og $F_2$.

Vi må ha: $\sum F_y = F_{1,y} + F_{2,y} + F_{B,y} = 0$.
Den minste kraften $F_B$ vil ikke ha noen x-komponent ($F_{B,x} = 0$).

$$F_{1,y} = 100 \sin(60^\circ) \approx 86.6 \, \text{N}$$
$$F_{2,y} = 140 \sin(-30^\circ) = -70.0 \, \text{N}$$
$$F_{B,y} = -(86.6 \, \text{N} - 70.0 \, \text{N}) = -16.6 \, \text{N}$$

Vektoren blir: $\vec{F}_B = [0, -16.6] \, \text{N}$.

### c) Finn kassens masse
Gitt at akselerasjonen i x-retning er $a = 2.0 \, \text{m/s}^2$. Vi bruker Newtons 2. lov: $\sum F_x = m \cdot a$.
$$\sum F_x = F_{1,x} + F_{2,x} = 100 \cos(60^\circ) + 140 \cos(-30^\circ)$$

```python
import numpy as np

# Definerer kreftene
F1_mag, F1_deg = 100.0, 60.0
F2_mag, F2_deg = 140.0, -30.0

# Dekomponering
F1x = F1_mag * np.cos(np.radians(F1_deg))
F1y = F1_mag * np.sin(np.radians(F1_deg))

F2x = F2_mag * np.cos(np.radians(F2_deg))
F2y = F2_mag * np.sin(np.radians(F2_deg))

# b) Beregne Fb_y for at y-komponentene skal nulle hverandre ut
Fby = -(F1y + F2y)
Fb = np.array([0, Fby])

print(f"Kraften Fb som sikrer horisontal bevegelse: {Fb.round(2)} N")

# c) Beregne masse m når akselerasjon a = 2.0 m/s^2
a_x = 2.0
sum_Fx = F1x + F2x
m = sum_Fx / a_x

print(f"Summen av krefter i x-retning: {sum_Fx:.2f} N")
print(f"Kassens masse m: {m:.2f} kg")
```
