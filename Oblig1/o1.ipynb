### a)
Dekomponer vektorene i sine respektive $x$- og
$y$-komponenter, og skriv derreter vektorene på formen $\overrightarrow{v}=v_x \hat{i} + v_y \hat{j}$,
eventuelt $\overrightarrow{v} = [v_x,v_y]$

```python
from numpy import cos, sin, radians
from IPython.display import Markdown

class Vector:
    def __init__(self,name:str, x:float|int,y:float|int):
        self.name:str = name
        self.x:float|int = x
        self.y:float|int = y

    @staticmethod
    def rename(name:str,vector:Vector)->Vector:
        vector.name = name
        return vector

    @staticmethod
    def lenDeg(name:str, length:float|int, angleDeg:float|int) -> Vector:
        x = length * cos(radians(angleDeg))
        y = length * sin(radians(angleDeg))
        return Vector(name,x,y)

    def display1(self):
        if(self.y<0):
            display(Markdown(f"$\\overrightarrow{{{self.name}}}= {self.x}\\hat{{i}}{self.y}\\hat{{j}}$"))
        else:
            display(Markdown(f"$\\overrightarrow{{{self.name}}}= {self.x}\\hat{{i}}+{self.y}\\hat{{j}}$"))

    def display2(self):
        display(Markdown(f"$\\overrightarrow{{{self.name}}}=[{self.x}, {self.y}]$"))

    def __add__(self:Vector,other:Vector) -> Vector:
        return Vector(f"{self.name}+{other.name}", self.x+other.x, self.y+other.y)
    
    def __sub__(self:Vector,other:Vector) -> Vector:
        return Vector(f"{self.name}-{other.name}", self.x-other.x, self.y-other.y)
    
    def __mul__(self:Vector,other:Vector) -> float|int:
        return self.x*other.x + self.y*other.y
    
    def __pow__(self:Vector,other:Vector) -> float|int:
        return self.x*other.y - self.y*other.x
    
    def __neg__(self:Vector) -> Vector:
        return Vector(f"-{self.name}",-self.x,-self.y)
```

```python
A=Vector.lenDeg("A",100,30)
A.display1()
A.display2()
```

## Oppgave 2

I denne oppgaven skal vi se nærmere på prosjektilbevegelse, hvor en gjenstand
kastes/skytes skrått ut i fra en startposisjon $(x_0, z_0) = (0, 0)$ ved tiden $t_a = 0$.
Vi forutsetter at prosjektilets startfart v0 og retning θ er kjent.
Posisjonen ⃗r(t) til prosjektilet kan beskrives av følgende vektorlikning:

$$\overrightarrow{r}(t) = (v_0 \cos{(\theta)}t)\hat{i} + \bigg(v_0\sin{(\theta)}t - \frac{1}{2}gt^2 \bigg)\hat{k} $$

### a)

Antar vi bare trenger høyderedning, der det er snakk om en relasjon av tid og posisjon, og der eneste kraftvektor er $g$.

forekler derfor reknestykke til:

$$\overrightarrow{r}(t) = v_0\sin{(\theta)}t - \frac{1}{2}gt^2$$

Deretter finner vi tiden der $\overrightarrow{r}(t) = 0$

$$0 = v_0\sin{(\theta)}t - \frac{1}{2}gt^2$$

$$\frac{1}{2}gt^2 = v_0\sin{(\theta)}t$$
$$gt^2 = 2v_0\sin{(\theta)}t$$
$$t^2 = \frac{2v_0\sin{(\theta)}t}{g}$$
$$t^2 = \frac{2v_0\sin{(\theta)}t}{g}$$

Dersom $t=0$ så er tiden kjent. Fortsetter derfor med $t\neq0$

$$t^2 = \frac{2v_0\sin{(\theta)}t}{g}$$
$$t = \frac{2v_0\sin{(\theta)}}{g}$$

Ser nå at jeg har en formel som ligner mistenkelig på det jeg skulle ende opp med,
og sier meg fornøyd.


### b)

Vi finner først et uttrykk for $t$ fra x-komponenten:  
$$x = v_0 \cos(\theta)t \Rightarrow t = \frac{x}{v_0 \cos(\theta)}$$

Dette settes inn i uttrykket for $\overrightarrow{z}(t)$:  
$$\overrightarrow{z(x)} = v_0 \sin(\theta) \left( \frac{x}{v_0 \cos(\theta)} \right) - \frac{1}{2}g \left( \frac{x}{v_0 \cos(\theta)} \right)^2$$

Som forenkles til:  
$$\overrightarrow{z(x)} = \tan(\theta)x - \frac{gx^2}{2v_0^2 \cos^2(\theta)}$$

### c)

Maksimal høyde nås når den vertikale hastigheten $v_z(t) = 0$:  
$$v_z(t) = \frac{dz}{dt} = v_0 \sin(\theta) - gt = 0 \Rightarrow t_{topp} = \frac{v_0 \sin(\theta)}{g}$$

Dette er nøyaktig halvparten av den totale flytiden $t_{fly}$.

### d) 
Setter $t_{topp}$ inn i $z(t)$:  
$$H = v_0 \sin(\theta) \left( \frac{v_0 \sin(\theta)}{g} \right) - \frac{1}{2}g \left( \frac{v_0 \sin(\theta)}{g} \right)^2 = \frac{v_0^2 \sin^2(\theta)}{2g}$$

### e, f, g)

```python
import numpy as np
import matplotlib.pyplot as plt

def prosjektil_simulering(v0, theta_deg, m=0.43):
    g = 9.81
    theta = np.radians(theta_deg)
    
    # Beregner flytid for å definere tidsaksen
    t_fly = (2 * v0 * np.sin(theta)) / g
    t = np.linspace(0, t_fly, 500)
    
    # Posisjon over tid
    x = v0 * np.cos(theta) * t
    z = v0 * np.sin(theta) * t - 0.5 * g * t**2
    
    # Hastighetskomponenter og total fart for energi
    vx = v0 * np.cos(theta)
    vz = v0 * np.sin(theta) - g * t
    v_fart = np.sqrt(vx**2 + vz**2)
    
    # Energi
    Ek = 0.5 * m * v_fart**2
    Ep = m * g * z
    Etot = Ek + Ep
    
    # Plotting
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Bane-plot
    ax1.plot(x, z, color='blue', label='Bane')
    ax1.set_title(f'Prosjektilbane (v0={v0}m/s, θ={theta_deg}°)')
    ax1.set_xlabel('Distanse (m)')
    ax1.set_ylabel('Høyde (m)')
    ax1.grid(True)
    
    # Energi-plot
    ax2.plot(t, Ek, label='Kinetisk energi (Ek)')
    ax2.plot(t, Ep, label='Potensiell energi (Ep)')
    ax2.plot(t, Etot, '--', label='Total mekanisk energi', color='black')
    ax2.set_title('Energi som funksjon av tid')
    ax2.set_xlabel('Tid (s)')
    ax2.set_ylabel('Energi (J)')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    print(f"Maks høyde: {np.max(z):.2f} m")
    print(f"Rekkevidde: {np.max(x):.2f} m")
    print(f"Total flytid: {t_fly:.2f} s")

# Kjører simulering for v0 = 30 m/s og vinkel 20 grader
prosjektil_simulering(30, 20)
```

# Oppgave 3: Krefter på en kasse

En kasse med masse $m$ befinner seg på et friksjonsfritt horisontalt underlag. To krefter virker på kassen:
* $F_1 = 100.0 \, \text{N}$ med vinkel $60.0^\circ$ over horisontalen.
* $F_2 = 140.0 \, \text{N}$ med vinkel $-30.0^\circ$ (under horisontalen).

### a) Frilegemediagram
Et frilegemediagram for kassen vil inneholde:
1. Tyngdekraften ($G = mg$) pekende nedover.
2. Normalkraften ($F_N$) fra underlaget pekende oppover.
3. Kraften $F_1$ pekende på skrå opp mot høyre.
4. Kraften $F_2$ pekende på skrå ned mot høyre.



### b) Finn minste kraft $F_B$ for horisontal bevegelse
For at kassen skal bevege seg rent horisontalt (kun langs x-aksen), må summen av kreftene i vertikal retning ($y$-aksen) være lik null. Vi ser bort fra tyngdekraft og normalkraft her, da oppgaven spør om hvilken *ekstra* kraft $F_B$ som må til for å kansellere ut de vertikale komponentene til $F_1$ og $F_2$.

Vi må ha: $\sum F_y = F_{1,y} + F_{2,y} + F_{B,y} = 0$.
Den minste kraften $F_B$ vil ikke ha noen x-komponent ($F_{B,x} = 0$).

$$F_{1,y} = 100 \sin(60^\circ) \approx 86.6 \, \text{N}$$
$$F_{2,y} = 140 \sin(-30^\circ) = -70.0 \, \text{N}$$
$$F_{B,y} = -(86.6 \, \text{N} - 70.0 \, \text{N}) = -16.6 \, \text{N}$$

Vektoren blir: $\vec{F}_B = [0, -16.6] \, \text{N}$.

### c) Finn kassens masse
Gitt at akselerasjonen i x-retning er $a = 2.0 \, \text{m/s}^2$. Vi bruker Newtons 2. lov: $\sum F_x = m \cdot a$.
$$\sum F_x = F_{1,x} + F_{2,x} = 100 \cos(60^\circ) + 140 \cos(-30^\circ)$$

```python
import numpy as np

# Definerer kreftene
F1_mag, F1_deg = 100.0, 60.0
F2_mag, F2_deg = 140.0, -30.0

# Dekomponering
F1x = F1_mag * np.cos(np.radians(F1_deg))
F1y = F1_mag * np.sin(np.radians(F1_deg))

F2x = F2_mag * np.cos(np.radians(F2_deg))
F2y = F2_mag * np.sin(np.radians(F2_deg))

# b) Beregne Fb_y for at y-komponentene skal nulle hverandre ut
Fby = -(F1y + F2y)
Fb = np.array([0, Fby])

print(f"Kraften Fb som sikrer horisontal bevegelse: {Fb.round(2)} N")

# c) Beregne masse m når akselerasjon a = 2.0 m/s^2
a_x = 2.0
sum_Fx = F1x + F2x
m = sum_Fx / a_x

print(f"Summen av krefter i x-retning: {sum_Fx:.2f} N")
print(f"Kassens masse m: {m:.2f} kg")
```

# Oppgave 4: Numerisk løsning av fall med luftmotstand

Vi skal studere et objekt med masse $m = 3.0 \, \text{kg}$ som faller fra ro. Objektet påvirkes av tyngdekraften og en luftmotstand som er proporsjonal med farten: $F_D = -kv$.

Den differensielle ligningen for bevegelsen er:
$$\frac{dv}{dt} = -g - \frac{k}{m}v$$

Vi bruker **Eulers metode** for å finne farten numerisk. Metoden går ut på å tilnærme den neste verdien for farten ved hjelp av den nåværende akselerasjonen og et lite tidsskritt $\Delta t$:
1. Beregn akselerasjonen: $a_i = -g - \frac{k}{m}v_i$
2. Oppdater farten: $v_{i+1} = v_i + a_i \cdot \Delta t$
3. Oppdater tiden: $t_{i+1} = t_i + \Delta t$

Vi bruker verdiene:
* $g = 9.81 \, \text{m/s}^2$
* $k = 1.0 \, \text{kg/s}$
* $m = 3.0 \, \text{kg}$
* $\Delta t = 0.25 \, \text{s}$

Terminalfarten ($v_t$) inntreffer når akselerasjonen er null:
$0 = -g - \frac{k}{m}v_t \Rightarrow v_t = -\frac{mg}{k} = -\frac{3.0 \cdot 9.81}{1.0} = -29.43 \, \text{m/s}$.

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_luftmotstand(m=3.0, k=1.0, dt=0.25, t_slutt=5.0):
    g = 9.81
    
    # Lister for å lagre resultater
    t_verdier = [0]
    v_verdier = [0] # Starter fra ro (v=0)
    
    t = 0
    v = 0
    
    # Eulers metode
    while t < t_slutt:
        a = -g - (k/m) * v
        v = v + a * dt
        t = t + dt
        
        t_verdier.append(t)
        v_verdier.append(v)
        
    return t_verdier, v_verdier

# Beregn verdier
t_euler, v_euler = euler_luftmotstand()

# Teoretisk terminalfart for sammenligning
v_terminal = -(3.0 * 9.81) / 1.0

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t_euler, v_euler, 'ro-', label=f'Euler (dt=0.25s)')
plt.axhline(y=v_terminal, color='k', linestyle='--', label=f'Terminalfart ({v_terminal:.2f} m/s)')

plt.title('Numerisk løsning av fall med luftmotstand (Eulers metode)')
plt.xlabel('Tid (s)')
plt.ylabel('Fart (m/s)')
plt.legend()
plt.grid(True)
plt.show()

# Skriv ut verdiene etter 5 sekunder
print(f"Fart etter 5 sekunder (numerisk): {v_euler[-1]:.2f} m/s")
```

# Oppgave 5 og 6: Kollisjoner

Vi har to vogner på en friksjonsfri bane:
* Vogn 1: $m_1 = 1.20 \, \text{kg}$, $v_1 = 5.0 \, \text{m/s}$ (mot høyre).
* Vogn 2: $m_2 = 0.50 \, \text{kg}$, $v_2 = -3.0 \, \text{m/s}$ (mot venstre).

### Oppgave 5: Elastisk støt
I et fullstendig elastisk støt er både bevegelsesmengden ($p$) og den kinetiske energien ($K$) bevart.
1. Bevaring av bevegelsesmengde: $m_1 v_1 + m_2 v_2 = m_1 u_1 + m_2 u_2$
2. Bevaring av kinetisk energi: $\frac{1}{2}m_1 v_1^2 + \frac{1}{2}m_2 v_2^2 = \frac{1}{2}m_1 u_1^2 + \frac{1}{2}m_2 u_2^2$



### Oppgave 6: Fullstendig uelastisk støt
I et uelastisk støt henger vognene sammen etter kollisjonen og beveger seg med en felles fart $v_{felles}$. Kun bevegelsesmengden er bevart.
$$m_1 v_1 + m_2 v_2 = (m_1 + m_2) v_{felles}$$
$$v_{felles} = \frac{m_1 v_1 + m_2 v_2}{m_1 + m_2}$$

Kinetisk energi før støtet: $K_{før} = \frac{1}{2}m_1 v_1^2 + \frac{1}{2}m_2 v_2^2$
Kinetisk energi etter støtet: $K_{etter} = \frac{1}{2}(m_1 + m_2) v_{felles}^2$

```python
import numpy as np

# Gitte verdier
m1 = 1.20
v1 = 5.0
m2 = 0.50
v2 = -3.0

# Oppgave 5: Elastisk støt (Formler for sluttfart u1 og u2)
u1 = ((m1 - m2) / (m1 + m2)) * v1 + ((2 * m2) / (m1 + m2)) * v2
u2 = ((2 * m1) / (m1 + m2)) * v1 + ((m2 - m1) / (m1 + m2)) * v2

print("--- Oppgave 5: Elastisk støt ---")
print(f"Fart vogn 1 etter støt: {u1:.2f} m/s")
print(f"Fart vogn 2 etter støt: {u2:.2f} m/s")

# Oppgave 6: Uelastisk støt
v_felles = (m1 * v1 + m2 * v2) / (m1 + m2)

# Kinetisk energi
K_for = 0.5 * m1 * v1**2 + 0.5 * m2 * v2**2
K_etter = 0.5 * (m1 + m2) * v_felles**2
delta_K = K_etter - K_for

print("\n--- Oppgave 6: Uelastisk støt ---")
print(f"Felles sluttfart: {v_felles:.2f} m/s")
print(f"Kinetisk energi før: {K_for:.2f} J")
print(f"Kinetisk energi etter: {K_etter:.2f} J")
print(f"Endring i kinetisk energi (tap): {delta_K:.2f} J")
```

# Oppgave 7: Statikk - Kran i likevekt

En kran er festet i en aksel (origo) og holdes oppe av en kabel. Vi har følgende data:
* Kranens tyngde: $W_k = 15000 \, \text{N}$
* Lastens tyngde: $W_p = 11000 \, \text{N}$
* Kranens lengde: $L = 16.0 \, \text{m}$
* Massesenter for kranen: $d_{cm} = 7.0 \, \text{m}$ fra akselen.
* Vinkel kran: $\theta = 55^\circ$ over horisontalplanet.
* Kabelen danner en vinkel på $25^\circ$ med selve kranen.
* Kabelen er festet $3.0 \, \text{m}$ fra toppen av kranen (altså $13.0 \, \text{m}$ fra akselen).

### (a) Frilegemediagram
Et frilegemediagram for kranen inkluderer:
1. Tyngdekraften fra kranen ($W_k$) som virker vertikalt ned fra massesenteret ($7.0 \, \text{m}$ ute).
2. Tyngdekraften fra lasten ($W_p$) som virker vertikalt ned fra tuppen ($16.0 \, \text{m}$ ute).
3. Snordraget ($T$) fra kabelen som virker i $25^\circ$ vinkel relativt til kranen ($13.0 \, \text{m}$ ute).
4. Reaksjonskrefter i akselen ($R_x$ og $R_y$).



### (b) Beregning av snordrag (T)
Vi bruker momentlikevekt om akselen ($\sum \tau = 0$) for å eliminere reaksjonskreftene i festepunktet.
Mot klokka regnes som positivt:
$$\sum \tau = \tau_{kabel} - \tau_{kran} - \tau_{last} = 0$$

* Momentarm for kranens tyngde: $r_k = 7.0 \cdot \cos(55^\circ)$
* Momentarm for lastens tyngde: $r_p = 16.0 \cdot \cos(55^\circ)$
* Moment fra kabelen: $\tau_{kabel} = T \cdot \sin(25^\circ) \cdot 13.0$

Ligningen blir:
$$T \cdot \sin(25^\circ) \cdot 13.0 = W_k \cdot 7.0 \cos(55^\circ) + W_p \cdot 16.0 \cos(55^\circ)$$

```python
import numpy as np

# Konstanter
W_k = 15000
W_p = 11000
theta_kran = np.radians(55)
vinkel_kabel_rel = np.radians(25)
L_feste = 13.0 # Kabelen er festet 13m fra akselen

# Moment fra tyngdekreftene (kraft * horisontal avstand)
moment_kran = W_k * 7.0 * np.cos(theta_kran)
moment_last = W_p * 16.0 * np.cos(theta_kran)
total_moment_ut = moment_kran + moment_last

# Snordrag T
# T * sin(vinkel_rel) * arm = total_moment_ut
T = total_moment_ut / (L_feste * np.sin(vinkel_kabel_rel))

print(f"Snordraget i kabelen (T): {T:.2f} N")

# (c) Finn kraften fra akselen (valgfritt men inkludert)
# Sum Fx = 0 => Rx - T_x = 0
# T sin(55-25) er ikke helt riktig, vi må bruke vinkler relativt til x-aksen
vinkel_T_x = np.radians(55 + 25) # Kabelen drar "brattere" enn kranen
Tx = T * np.cos(vinkel_T_x)
Ty = T * np.sin(vinkel_T_x)

Rx = Tx # Siden sum Fx = 0
Ry = (W_k + W_p) - Ty # Siden sum Fy = 0

R_total = np.sqrt(Rx**2 + Ry**2)
print(f"Kraften fra akselen på kranen: {R_total:.2f} N")
```

# Oppgave 8: Rotasjonsdynamikk - Romaskin

Vi har et system med en snor som trekkes over tre trinser til en justerbar last $m_L$.
* Masse per trinse: $m_T = 1.0 \, \text{kg}$
* Radius per trinse: $r = 0.15 \, \text{m}$
* Treghetsmoment for en massiv skive (trinse): $I = \frac{1}{2} m_T r^2$
* Lastens masse: $m_L$ (variabel)

Når brukeren drar i snora med en kraft $F_{snor}$, vil denne kraften akselerere både lasten lineært og trinsene rotasjonelt.

### (a) Akselerasjon uten last ($m_L = 0$)
Hvis vi antar at snora er masseløs og ikke glir, vil kraften $F_{snor}$ gå til å akselerere de tre trinsene. 
Momentet på én trinse er $\tau = \Delta F \cdot r$, der $\Delta F$ er forskjellen i snordrag over trinsen. For tre trinsere må vi overvinne tregheten til alle tre.
Forbindelsen mellom lineær akselerasjon ($a$) og vinkelakselerasjon ($\alpha$) er $a = \alpha \cdot r$.

Newtons 2. lov for rotasjon: $\sum \tau = I_{total} \cdot \alpha$.
Siden kraften virker på ytterkanten av trinsene, har vi $F_{snor} \cdot r = (3 \cdot I) \cdot \frac{a}{r}$.
Dette gir:
$$F_{snor} = 3 \cdot \left(\frac{1}{2} m_T r^2\right) \cdot \frac{a}{r^2} = \frac{3}{2} m_T a$$
Akselerasjonen blir da:
$$a = \frac{2 \cdot F_{snor}}{3 \cdot m_T}$$

### (b) Effekt av lasten $m_L$
Når vi legger til en last $m_L$, må kraften også akselerere denne massen lineært, i tillegg til å overvinne tyngdekraften hvis lasten løftes vertikalt (avhengig av maskinens design, her antar vi lineær akselerasjon av hele systemet).
Den totale "effektive massen" $m_{eff}$ som brukeren føler er:
$$m_{eff} = m_L + \frac{3}{2} m_T$$

```python
import numpy as np

# Gitte verdier
m_t = 1.0  # kg per trinse
r = 0.15   # m
F_snor = 150.0  # Antatt trekkraft i Newton for eksempelet

# a) Beregn akselerasjon a når m_L = 0
# Basert på utledningen: a = F / (1.5 * m_t) for 3 trinser
a_uten_last = F_snor / (1.5 * m_t)

print(f"--- Oppgave 8 ---")
print(f"Akselerasjon uten last (F={F_snor}N): {a_uten_last:.2f} m/s^2")

# b) Hva skjer hvis vi legger til en last m_L = 10 kg?
m_l = 10.0
a_med_last = F_snor / (m_l + 1.5 * m_t)

print(f"Akselerasjon med 10kg last: {a_med_last:.2f} m/s^2")

# Treghetsmoment for én trinse
I = 0.5 * m_t * r**2
print(f"Treghetsmoment per trinse: {I:.4f} kg*m^2")
```
