### a)
Dekomponer vektorene i sine respektive $x$- og
$y$-komponenter, og skriv derreter vektorene p√• formen $\overrightarrow{v}=v_x \hat{i} + v_y \hat{j}$,
eventuelt $\overrightarrow{v} = [v_x,v_y]$

```python
from numpy import cos, sin, radians
from IPython.display import Markdown

class Vector:
    def __init__(self,name:str, x:float|int,y:float|int):
        self.name:str = name
        self.x:float|int = x
        self.y:float|int = y

    @staticmethod
    def lenDeg(name:str, length:float|int, angleDeg:float|int) -> Vector:
        x = length * cos(radians(angleDeg))
        y = length * sin(radians(angleDeg))

        return Vector(name,x,y)

    def display1(self):
        display(Markdown(f"$\\overrightarrow{{{self.name}}}= {self.x}\\hat{{i}} + {self.y}\\hat{{J}}$"))

    def display2(self):
        display(Markdown(f"$\\overrightarrow{{{self.name}}}=[{self.x}, {self.y}]$"))


A=Vector.lenDeg("A",100,30)
B=Vector.lenDeg("B",80,30+90)
C=Vector.lenDeg("C",40,53+90+90)

A.display1()
A.display2()
B.display1()
B.display2()
C.display1()
C.display2()
```
